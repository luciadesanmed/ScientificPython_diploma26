#### **Exercise 1:**
Create a function returning the first N terms of the tailor series of the function without using the sympy series function.

```python
import sympy as sp
import numpy as np

#Function to return first n terms of the taylor series:
def taylor(func,nterms,value):
 x=sp.symbols('x')
 f=func(x)
 terms=0
    
 for i in range(nterms):
  der=sp.diff(f,x,i)
  serie=(der/sp.factorial(i))*(x-value)**i
  terms=terms+serie
 return terms

taylor(sp.sin,5,0)
```

#### **Exercise 2:**

Write a matrix multiplication function using lists ("manually", using the matrix multiplication formula) and another one using numpy arrays (call matmul function). Time them both. 

```python
import numpy as np

#Matrix multiplication using lists (using the matrix multiplication formula):
#Considering square matrices:
def matrix_mult_list(a,b):
 c=np.zeros((len(a),len(b)))
 for i in range(len(a)):
  for j in range(len(a)):
   fila=a[i]
   col=[b[k][j] for k in range(len(a))]
   c[i][j]=sum(fila[k]*col[k] for k in range(len(a)))
 return c

#Using numpy arrays:
def matrix_mult_numpy(a,b):
 c=np.matmul(a,b)
 return c

a=[[1,2],[3,2]]
b=[[2,2],[3,6]]
print('Multiplication with lists:', matrix_mult_list(a,b))

print('Multiplication with numpy:',matrix_mult_numpy(a,b))
```

#### **Exercise 4:**
Plot the recaman sequence in the following way:

Add a semi-circular arc from $a_{n}$ to $a_{n-1}$.

Alternate the arcs to be above and below the x-axis according to whether
    n is even or odd.

```python
import numpy as np
from matplotlib import pyplot as plt
#Recaman sequence:
#Add a semi-circular arc from an to an-1:
#Alternate the arcs to be above and below the x-axis according to whether n is even or odd:
def recaman(max_terms):
    exist = set()
    seq=list()
    n = 0 
    a = 0
    while n < max_terms:
        diff=a-n
        if diff > 0 and diff not in exist:
            a = diff
        else:
            a = a + n
        exist.add(a)
        seq.append(a)
        n += 1
    return seq

n=26
seq=recaman(n)

for i in range(n-1):
 r=(seq[i+1]-seq[i])/2
 minmax=[seq[i+1],seq[i]]
 x=np.linspace(np.min(minmax),np.max(minmax),100)
 c=(seq[i]+seq[i+1])/2
 y=np.sqrt(r**2-(x-c)**2)
 if (i+1)%2==0:
  plt.plot(x,y,'m')
 else:
  plt.plot(x,-y,'m')  
```

#### **Exercise 6:**

Plot a nice-looking 3-D torus. Do this 3 times with different styles. 

```python
#Nice-looking 3D torus:
import matplotlib as mp
from mpl_toolkits import mplot3d

c=10
a=2
u = np.linspace(0,2*np.pi,200)
v = np.linspace(0,2*np.pi,200)
U,V=np.meshgrid(u,v)

X = (c+a*np.cos(V))*np.cos(U)
Y = (c+a*np.cos(V))*np.sin(U)
Z = a*np.sin(V)

fig = plt.figure(figsize=(15,10))
ax=fig.add_subplot(131,projection='3d')
ax1=fig.add_subplot(132,projection='3d')
ax2=fig.add_subplot(133,projection='3d')

#Different styles:
#Surface:
ax.plot_surface(X,Y,Z,cmap='prism')
#Wireframe:
ax1.plot_wireframe(X,Y,Z)
#Contour:
ax2.contour3D(X,Y,Z)
```

#### **Exercise 7:**

Repeat the curve fitting example (from SciPy lecture) with the following modifications:

 - choose some other function
 - first write the data to a file, then read it (with numpy, for example)
 - produce the picture that has axes, name and legend


```python
from scipy import optimize

np.random.seed(0)

# Let's generate fake data that we can "fit" then 
def f(x, a, phi):
    return np.tan(a*x-phi)

#random makes it look like experimental data
x = np.linspace(0, 4, 50)
y = f(x, 1.3, 0.9) + .1*np.random.normal(size=50)

# Now fit the model: the parameters omega and phi can be found in the
# `params` vector
params, params_cov = optimize.curve_fit(f, x, y)

print(params)
#Save the data to a file and then read it:
np.savetxt('data.txt',[x,y])
#Open
dat=np.loadtxt('data.txt')

# plot the data and the fitted curve
ax=plt.axes()
ax.plot(dat[0,:], dat[1,:], 'bx')
ax.plot(dat[0,:],f(dat[0,:],1.3,0.9),'g-')
ax.plot(dat[0,:], f(dat[0,:], *params), 'r-')
ax.set_title('Figure',fontsize = 'large',fontweight='bold')
ax.set_ylabel('$f(x)$',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'heavy')
ax.set_xlabel('$x$',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
ax.legend(('Experimental data','Function','Model fit'),loc='upper right', shadow=True)
```

Integrate the function `x**2 + 10*np.sin(x)` on the interval `[-10,10]`.
Find a way to make a plot like in textbooks where you color the area under the curve that you integrate.
Now integrate the same function symbolically using `sympy`, then put borders into it and compare the results.

```python
import numpy as np
import sympy as sp
import matplotlib as mp
from matplotlib import pyplot as plt
#Integrate the function, find a way to make a plot like in textbooks where you color the area under the curve that you integrate:
x=np.linspace(-10,10,100)
y=x**2 + 10*np.sin(x)
c=np.zeros(len(x))
plt.plot(x,y,'-r')
plt.plot(x,c)
plt.fill_between(x,c,y)
plt.plot(x,np.zeros(100))

#Symbolically using sympy:
x=sp.symbols('x')
sp.integrate(x**2+10*sp.sin(x))
#Adding borders to it:
sp.integrate(x**2+10*sp.sin(x),(x,-10,10))
```
